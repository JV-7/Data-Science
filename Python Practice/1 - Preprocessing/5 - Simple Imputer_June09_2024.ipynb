```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.compose import make_column_transformer
```

```python
miles = pd.DataFrame({'farthest_run_mi': [50,62, np.nan,100,26,13,31,50, None]})
miles
```

```python
miles.isnull().sum()
```

```python
imp_mean = SimpleImputer(strategy='mean')
imp_mean.fit_transform(miles)
```

```python
imp_median = SimpleImputer(strategy='median')
imp_median.fit_transform(miles)
```

```python
imp_mode = SimpleImputer(strategy='most_frequent')
imp_mode.fit_transform(miles)
```

```python
imp_constant = SimpleImputer(strategy='constant', fill_value=20)
imp_constant.fit_transform(miles)
```

```python
names = pd.DataFrame({'Name': ['ryan', 'nolan', 'honus', 'wagner', 'ruth', np.nan, 'Jaya', 'JV', np.nan]})
names
```

```python
imp_constants_cat = SimpleImputer(strategy='constant', fill_value= 'JV')
imp_constants_cat.fit_transform(names)
```

## to see where the simple imputer worked

```python
imp_mean_marked = SimpleImputer(strategy='mean',add_indicator=True)
miles_imputed = imp_mean_marked.fit_transform(miles)
miles_imputed
imputed_columns  =  imp_mean_marked.get_feature_names_out(input_features=miles.columns)
imputed_columns_df = pd.DataFrame(miles_imputed, columns=imputed_columns)

imputed_columns_df
```

```python
Name_Miles = pd.concat([names, miles], axis=1)
Name_Miles
```

```python
imp_constants_cat = SimpleImputer(strategy='constant', fill_value= 'JV')
imp_mean = SimpleImputer(strategy='mean')

ct = make_column_transformer(
    (imp_constants_cat, ['Name']),
    (imp_mean,['farthest_run_mi'])
).set_output(transform='pandas')

# ct.set_output(transform='pandas')
ct_ft = ct.fit_transform(Name_Miles)
ct_df = pd.DataFrame(ct_ft)
ct_df.columns = ['A', 'B']
ct_df
```
