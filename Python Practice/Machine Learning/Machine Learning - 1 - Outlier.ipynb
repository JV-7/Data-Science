```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

```

  <b>  np.random.normal(loc=170, scale=5, size=1000): Generates 1000 random numbers from a normal distribution with a mean of 170 and a standard deviation of 5

```python
np.random.seed(100)

height_without_outliers = np.random.normal(loc=170, scale=5, size=10)
height_with_outliers = np.random.randint(low=10, high=50, size=10)

height_with_outliers = np.append(height_with_outliers, [100, 150, 180, 200, 210])

print(height_without_outliers)
print(height_with_outliers)
```

### with Matplotlib

```python
fig, axis = plt.subplots(1, 2, figsize=(10,5))

#height_without_outliers - First figure
axis[0].boxplot([height_without_outliers], vert=False)

#height_without_outliers - second figure
axis[1].boxplot(height_with_outliers, vert=False)

```

### with seaborn

```python
fig, axis = plt.subplots(1, 2, figsize=(10, 5))

sns.boxplot(x=height_without_outliers, ax=axis[0])
sns.boxenplot(x=height_with_outliers, ax=axis[1])
```
